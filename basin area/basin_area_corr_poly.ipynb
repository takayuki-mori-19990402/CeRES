```python
!pip -q install japanize_matplotlib pyvis==0.1.9
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
import japanize_matplotlib

from sklearn.datasets import fetch_california_housing

import networkx as nx
from pyvis.network import Network
from IPython.display import HTML
from sklearn.linear_model import LinearRegression
```

```python
a = "Area_2013"
b = "Area_calc"
c = "Area_2023"
```

```python
file_d = r"C:\Users\HONGO-23\Desktop\data\Irrigation\集水域\DOWN\DOWN_STREAM_diff.csv" 
df_d = pd.read_csv(file_d)
df_d.head()
```

```python
# 散布図を描画
plt.figure(figsize=(5, 5))  # figsizeを設定
plt.scatter(df_d[a], df_d[b], zorder=2, s = 100, c="magenta")
plt.xlabel(a)
plt.ylabel(b)
plt.xlim(-5,145)
plt.ylim(-5,145)

# LinearRegressionモデルを作成し、R^2を計算
model = LinearRegression()
model.fit(df_d[[a]], df_d[b])
r_squared = model.score(df_d[[a]], df_d[b])

# R^2を表示
plt.title('R2 = {:.2f}'.format(r_squared), fontsize = 15)

# 斜めの線を追加
plt.axline((0, 0), (100, 100), color='black', zorder=1)

# 図の背景を白くする
plt.gca().set_facecolor('white')

# 枠を黒くする
plt.gca().spines['top'].set_color('black')
plt.gca().spines['right'].set_color('black')
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['left'].set_color('black')
plt.grid(color='lightgray')
plt.text(23, 85, 'BCRJ15'.format(r_squared), fontsize=13)
plt.text(38, 69, 'BCKR4'.format(r_squared), fontsize=13)
plt.text(45, 16, 'BCKR2'.format(r_squared), fontsize=13)
plt.text(80, 56, 'BCRJ16ki'.format(r_squared), fontsize=13)

plt.savefig(r"C:\Users\HONGO-23\Desktop\data\Irrigation\集水域\DOWN\down_1.jpeg",dpi=200)
plt.show()
```
